**4. Data Wrangling**

```python
df= df.drop(['weight', 'admission_type_id'],axis=1)
```

```python
df=df.drop(['discharge_disposition_id', 'admission_source_id'], axis=1)
```

```python
df=df.drop(['diag_2', 'diag_3'], axis=1)
```

```python
df=df.drop(['number_emergency', 'diag_1'], axis=1)
```

```python
df=df.drop(['number_diagnoses', 'max_glu_serum'], axis=1)
```

```python
df=df.drop(['A1Cresult', 'metformin'], axis=1)
```

```python
df=df.drop(['repaglinide', 'nateglinide'], axis=1)
```

```python
df
```

```python
lab_dummies = pd.get_dummies(df['num_lab_procedures'], drop_first = True)
lab_dummies
```

```python
df.drop(['num_lab_procedures'], axis=1, inplace=True)
```

```python
df = pd.concat([df, lab_dummies], axis=1)
```

```python
df
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df['medical_specialty'])
```

```python
alexa_countvectorizer.shape
```

```python
type(alexa_countvectorizer)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())  
```

```python
df.drop(['medical_specialty'], axis=1, inplace=True)
reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
df= pd.concat([df, reviews], axis=1)
```

```python
df
```

```python
X = df.drop(['encounter_id'],axis=1)
```

```python
X
```

```python
y = df['encounter_id']
```

**5. Build Model**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

**6. Evaluate Models** 

```python
y_predict_train = randomforest_classifier.predict(X_train)
sm = confusion_matrix(y_train, y_predict_train)
```

```python
sns.heatmap(sm, annot=True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
sm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(sm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
sm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(sm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```
